```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Ensure the 'duration' column is of string type
df['Duration'] = df['Duration'].astype(str)

# Extract the integer part from the 'duration' column
df['Duration'] = df['Duration'].str.extract(r'(\d+)').astype(float)

# Calculate the mean of the 'duration' column, ignoring NaN values
mean_duration = df['Duration'].mean()

# Replace NaN values in the 'duration' column with the mean
df['Duration'].fillna(mean_duration, inplace=True)

print(f"The mean of the duration column is: {mean_duration}")
print(df)
df.isnull().sum()
```

```python
# Replace NaN values in the 'Year' column with the value above it (forward fill)
df['Year'].fillna(method='ffill', inplace=True)

# Replace any remaining NaN values in the 'Year' column with the value below it (backward fill)
df['Year'].fillna(method='bfill', inplace=True)
df.isnull().sum()
```

```python
df['Genre'].fillna(method='bfill', inplace=True)
df.isnull().sum()
```

```python

# Replace null values in 'Rating' column with the mean
mean_rating = df['Rating'].mean()
df['Rating'].fillna(mean_rating, inplace=True)

# Now 'Rating' column should have no null values
print(df.isnull().sum())
```

```python
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')

avg_votes= df['Votes'].mean()
df['Votes'].fillna(avg_votes, inplace=True)
df['Votes'] = df['Votes'].astype(int)
df.isnull().sum()
```
